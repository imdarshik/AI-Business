# **Strategic Plan for Influencer Marketing on TikTok**

**Goal:** Leverage influencer marketing to boost TikTok's growth by identifying and collaborating with high-impact influencers across various categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("tiktok_users/TikTok profiles dataset (Public web data).csv", on_bad_lines = "skip")
df.head()
```

```python
cols_to_keep = ["timestamp", "account_id", "biography", "awg_engagement_rate", "comment_engagement_rate", "like_engagement_rate", "is_verified", "followers", "following", "likes", "videos_count"]
df = df[cols_to_keep]
df.duplicated().sum()
```

```python
df.info()
```

```python
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.describe().round(2)
```

```python
import torch
from transformers import pipeline

if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
    device = torch.device("cpu")
    print("Using CPU")

classifier = pipeline("zero-shot-classification", model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli", device=device)
```

```python
# Define categories and labels
categories = ["Entertainment", "Dance", "Pranks", "FItness/Sports", "Home Reno/DIY", "Beauty/Skincare", "Fashion", "Recipes/Cooking", "Life hacks/advice etc.", "Pets", "Outdoors"]
```

```python
# Apply classification
results = classifier(df['biography'].to_list(), categories, multi_label=True)
```

```python
tmp = pd.DataFrame(results)
tmp["account_id"] = df["account_id"]
tmp_expanded = tmp.explode(['labels', 'scores'])
tmp_pivot = tmp_expanded.pivot_table(index=['account_id', 'sequence'], columns='labels', values='scores', fill_value=0,)
tmp_pivot = tmp_pivot.reset_index()
df_temp = pd.concat([df, tmp_pivot.sort_values("account_id")[categories]], axis=1)
```

```python
df_temp = df_temp.drop(columns=["timestamp", "biography"])
df_temp.to_csv("tiktok.csv", index=False)
```

```python
df = pd.read_csv("tiktok.csv", on_bad_lines = "skip")
df.head()
```

```python
influencer_map = {
    "not": 0,
    "nano": 1,
    "micro": 2,
    "macro": 3,
    "mega": 4
}

influencer_rank = {
    0: [0, 1000],
    1: [1001, 10000],
    2: [10001, 100000],
    3: [100001, 1000000],
    4: [1000001, np.inf]
}

# Function to determine influencer type based on follower count
def determine_influencer_type(followers):
    for influencer, (low, high) in influencer_rank.items():
        if low <= followers <= high:
            return influencer
    return None
```

```python
# Calculate the category with the highest confidence score for each influencer
df['top_category'] = df[categories].idxmax(axis=1)

# Display the distribution of top categories
category_distribution = df['top_category'].value_counts()

# Plot the distribution of top categories
sns.barplot(category_distribution, orient="h")
plt.ylabel('Category')
plt.xlabel('Number of Influencers')
plt.title('Distribution of Influencers by Top Category')
plt.show()
```

```python
df["influencer_rank"] = df['followers'].apply(determine_influencer_type)
```

```python
influencer_rank_count = df["influencer_rank"].value_counts().sort_values(ascending=False)
# Plot the distribution of top categories
sns.barplot(influencer_rank_count, orient="h")
plt.ylabel('Category')
plt.xlabel('Number of Influencers')
plt.title('Distribution of Influencers by followers rank')
plt.show()
```

```python
# Filter to find top influencers
top_influencers = df[df['influencer_rank'] > 2].sort_values(by='awg_engagement_rate', ascending=False)

# Display the top 10 influencers
top_influencers.iloc[:,list(range(9))+[-1]].head(10).round(4)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR

# Define models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
    'SVR': SVR(),
}
```

```python
tmp = df[df['influencer_rank'] > 2].drop(columns=['account_id','influencer_rank','top_category']+categories)
tmp.head()
```

```python
X = tmp[tmp.columns[3:]]
y = tmp[tmp.columns[0]]
```

```python
# Initialize scaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
# Train and evaluate models
for name, model in models.items():
    model.fit(X, y)
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    print(f'{name}: MSE = {mse:.4f}, R^2 = {r2:.4f}')
```
